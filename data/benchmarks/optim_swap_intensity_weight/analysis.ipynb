```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
scores_info = pd.read_csv("results/scores.csv")
(scores_info["algorithm_converged"] == False).sum()
```

```python
results_info = pd.read_csv("results/results.csv")
sns.violinplot(
    data=results_info[
        results_info["algorithm_converged"]
        & (results_info["assignment_name"] == "max_intersect")
        & (results_info["metric_name"] == "f1score")
    ],
    x="spellmatch_intensity_weight",
    y="metric_value",
    inner="point",
)
plt.xlabel(r"$w_\mathrm{intensity}$")
plt.ylabel(r"$F_1$-score")
plt.ylim(-0.15, 1.15)
plt.show()
```
